```python
# Deleting rows with null values
df= df.dropna()
```

```python
df['label'].value_counts()
```

```python
def clean_text(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
# applying clean_text function to each of the data in the text column
df['clean_text']=df['text'].apply(clean_text)
```

```python
df.head()
```

```python
import nltk
nltk.download('stopwords')
stop_words = stopwords.words('english')
more_stopwords = ['u', 'im', 'c']   # can add more stopwords as desired
stop_words = stop_words + more_stopwords

def remove_stopwords(text):
    words=text.split(' ')
    words=[x for x in words if x not in stop_words]
    text=' '.join(words)
    return text

df['clean_text']=df['clean_text'].apply(remove_stopwords)
df.head()
```

```python
# Stemming is a text normalization process that reduces words to their root or base form.

stemmer =nltk.SnowballStemmer('english')

def stemm_text(text):
    text = ' '.join(stemmer.stem(word) for word in text.split(' '))
    return text
df['clean_text']=df['clean_text'].apply(stemm_text)
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
import tensorflow as tf
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(df['clean_text'])
y_train = df['label']
X_test = vectorizer.transform(df['clean_text'])
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state = 42)
```

```python
model = LogisticRegression(random_state = 42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(X_val)
acc = accuracy_score(y_val, y_pred)
print(f'acc:{acc}')
```
